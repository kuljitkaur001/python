# seperating vowels from consonents 

```python
str24 = "me good, you say "
vowels = ["a", "e", "i", "o","u"]
v_count = " "
v_count2 = 0
conso_count = " "
conso_count2 = 0



for i in str24 :
    if i in vowels:
        v_count = v_count+i
        v_count2 = v_count2+1
print("the vowels exited in the given string are = ",v_count)
print("no. of vowels exist are = ",v_count2)

for j in str24:
    if j not in vowels:
        conso_count = conso_count+j
        conso_count2 = conso_count2+1
        
print("\nconsonents are = ",conso_count)
print("no of consonents are = ",conso_count2)
```

```python
str28 = "mom"
n = len(str28)
print(n) #lenth = 3 
is_palin = True # now it is palindrome k we consider it 

for i in range(0,n): # range we use when have lenth, like first  index and last index 
    if str28[i]== str28[n-1]: # str[1] = m, str[n-1]= str[3-1] = str[2] = m , if  m=m , 
        n = n-1 # n= 2, n decreses here, by 1 , now n= 2 
    else:
        is_palin = False # if m != m , it mean, it was not palindrome we have considered above 

if is_palin: # logic part 2: 
    print("yes")
else:
    print("no")
    

```

# plindrome using fucntion 

```python
def palindrome(n):
    if n == n[: :-1]:
        print("yes babes, it is palindrome")

    else:
        print("baes , you are sorry it is not palindrome")


str29 = input("enter a string ")
palindrome(str29)
```
